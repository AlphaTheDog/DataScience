```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#reading csv file
df=pd.read_csv("matches.csv")
df.head()
```

```python
#DF's Columns
df.columns
```

```python
#DATA TYPES
df.dtypes
```

```python
#SUMMARY STATISTICS
df.describe()
```

```python
#Dropping irrelevant columns
columns_drop=['notes','match report']
df1=df.drop(columns_drop,axis=1)
df1.drop('attendance',axis=1,inplace=True)
```

```python
df1.drop('dist',axis=1,inplace=True)
df1.drop('comp',axis=1,inplace=True)
```

```python
#finding duplicate entries
df1.duplicated().sum()
```

```python
#finding null values
df1.isnull().sum()
```

```python
#Adding column total points dependin on column result
conditions=[(df1['result']=='L'),(df1['result']=='W'),(df1['result']=='D')]
values=[0,3,1]
df1['points']=np.select(conditions,values)
```

```python
df1.head()
```

```python
#EPL standings per season
table=df1.groupby(['season','team'])['points'].sum()
print("2021")
print(table[2021].sort_values(ascending=False))
print("2022")
print(table[2022].sort_values(ascending=False))
```

```python
df1.head()
```

```python
#unique formations used in the games
df1['formation'].unique()
```

```python
#teams that managed to record most shots on target
shots=df1.groupby(['season','team'])['sot'].sum()
print("2021")
print(shots[2021].sort_values(ascending=False))
print("2022")
print(shots[2022].sort_values(ascending=False))
```

```python
df1.head()
```

```python
#aggreggate on different features per season and team
agg=df1.groupby(['season','team'])[['gf','ga','pk']].sum()
print(agg)
```

```python
#number of matches officiated by each referee
df1.groupby('referee')['referee'].count().sort_values(ascending=False)
```

```python
df1.head()
```

```python
#number of penalties awarded and scored by each referee
df1.groupby('referee')[['pkatt','pk']].sum().sort_values(by='pkatt',ascending=False)
```

```python
#points gained by team against opposition
enc=df1.groupby(['team','opponent'])[['result','points']]
for (hometeam,awayteam),group in enc:
    group['points']=np.where(group['result']=="L",-3,3)
    total_points=group['points'].sum()
    #print(hometeam,awayteam)
    print(hometeam,total_points,awayteam)
```

```python
#formation that won the most points
form=df1.groupby('formation')['points'].sum()
print(form.sort_values(ascending=False))
```
